```python
from pyfurcation.bifurcation import *
import anhima
import numpy as np
import scipy.stats
import random
%matplotlib inline
```

```python
# simulate non-uniform variant positions
np.random.seed(42)

n_variants = 200
p = 0
pos = [p]
for i in range(n_variants-1):
    gap = int(np.abs(np.cos(i/100))*100)
    p += gap
    pos.append(p)
pos = np.array(pos)
```

```python
# simulate genotypes
n_samples = 100
start_haplotype = np.zeros(n_variants)
h = [start_haplotype]

for i in range(1, n_samples):
    hap = random.choice(h).copy()
    mutations = np.random.randint(0, n_variants, 3)
    for m in mutations:
        hap[m] = (hap[m]-1)**2
    h.append(hap)

haplotypes = np.array(h).T
print haplotypes.shape
```
